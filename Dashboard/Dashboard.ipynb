```python
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.common.by import By  
from selenium.webdriver.support.ui import WebDriverWait # type: ignore
from selenium.webdriver.support import expected_conditions as EC 

import requests
from time import sleep

import pandas as pd
```

```python
def click(driver: webdriver.Chrome, el_value: str, wait_time: float=0.5, by=By.XPATH):
    sleep(wait_time)
    button = driver.find_element(by=by, value=el_value)
    button.click()

def get_value(driver: webdriver.Chrome, el_value: str, wait_time: float=0.0, by=By.XPATH):
    sleep(wait_time)
    el = driver.find_element(by=by, value=el_value)
    return el.text.replace(',', '.')

def fill_input(driver: webdriver.Chrome, el_value: str, input_value: str, wait_time: float=0.0, by=By.XPATH):
    sleep(wait_time)
    input = driver.find_element(by=by, value=el_value)
    input.clear()
    input.send_keys(input_value)

def open_page(url: str, implicity_wait: float=10):
    chrome_options = webdriver.ChromeOptions()

    # chrome_options.add_argument(' headless')
    # chrome_options.add_argument(' —-no-sandbox')
    # chrome_options.add_argument(' —-disable-dev-shm-usage')

    # chrome_options.add_argument('--disable-dev-shm-usage')
    # chrome_options.add_argument('--headless')
    # chrome_options.add_argument('--start-maximized') 
    # chrome_options.add_argument('--start-fullscreen')
    # chrome_options.add_argument('--single-process')

    chrome_options.add_argument('window-size=1300,1400')
    # chrome_options.headless = True

    driver = webdriver.Chrome('chromedriver', options=chrome_options)
    driver.implicitly_wait(implicity_wait)

    driver.get(url)

    return driver
```
